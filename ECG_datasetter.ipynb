```python
def calculate_stats(data, start_time=525, window_size=270, step=30):
  stats = []
  window_half = window_size // 2

  for window_start in range(start_time, len(data), step):
    window_end = window_start + window_size
    window_indices = np.where((data[:, 0] >= window_start - window_half) &
                             (data[:, 0] <= window_start + window_half))[0]
    if len(window_indices) > 0:
      rr_intervals = data[window_indices, 1]
      mean_rr = np.mean(rr_intervals)
      median_rr = np.median(rr_intervals)
      stats.append((mean_rr, median_rr))

  return stats
```

```python
def calculate_percentile(data, start_time=525, window_size=270, step=30):
  percentiles = [5, 10, 25, 50, 75, 90, 95]

  stats = []
  window_half = window_size // 2

  for window_start in range(start_time, len(data[:, 0]), step):
    window_indices = np.where((data[:, 0] >= window_start - window_half) &
                             (data[:, 0] <= window_start + window_half))[0]
    if len(window_indices) > 0:
      window_data = data[window_indices, 1]
      quantiles = np.quantile(window_data, [0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
      stats.append(quantiles)
  return stats
```

```python


def calculate_hrv_features(data, start_time=525, window_size=270, step=30):

  stats = []
  window_half = window_size // 2

  for window_start in range(start_time, len(data), step):
    window_end = window_start + window_size
    window_indices = np.where((data[:, 0] >= window_start - window_half) &
                             (data[:, 0] <= window_start + window_half))[0]
    if len(window_indices) > 0:
      rr_intervals = data[window_indices, 1]
      sdnn = np.std(rr_intervals)
      rr_range = np.max(rr_intervals) - np.min(rr_intervals)
      rr_diffs = np.diff(rr_intervals)
      pnn50 = np.sum(np.abs(rr_diffs) > 0.05) / len(rr_diffs) * 100
      rmssd = np.sqrt(np.mean(rr_diffs ** 2))
      sdsd = np.std(rr_diffs)
      mad = np.mean(np.abs(rr_intervals - np.mean(rr_intervals)))
    stats.append((sdnn, rr_range, pnn50, rmssd, sdsd, mad))

  return stats

```

```python
from copy import deepcopy
```

```python
def detrend_rr(data, start_time=915, window_size=900, step=30):
  detrended_data = deepcopy(data)
  changed = set()
  previous_ind = 0
  start = start_time - window_size
  for window_start in range(start, len(data)):
    window_indices = np.where((data[:, 0] >= window_start) &
                             (data[:, 0] <= window_start + window_size))[0]
    if len(window_indices) > 0:
      if window_indices[-1] not in changed:
        changed.add(window_indices[-1])
        rr_intervals = data[window_indices, 1]
        mean_rr = np.mean(rr_intervals)
        if window_indices[-1] - previous_ind > 1:
          for i in range(previous_ind + 1, window_indices[-1] + 1):
            detrended_data[i][1] -= mean_rr

        else:
          detrended_data[window_indices[-1]][1] -= mean_rr
        previous_ind = window_indices[-1]
  return detrended_data
```

```python
from scipy.fft import fft, fftfreq
```

```python
def calculate_freq_features(data, start_time=105000, window_size=270, step=30, fs=200):

  stats = []
  window_size *= fs
  step *= fs
  window_half = window_size // 2
  amount_of_points = len(data)
  begin = 0
  end = 0

  for window_start in range(start_time, amount_of_points, step):
    window_end = window_start + window_size
    begin = window_start - window_half
    end = window_start + window_half
    rr_intervals = data[begin:end]
    #if len(window_indices) > 0:
      #rr_intervals = data[window_indices]
    fft_data = fft(rr_intervals)
    f = fftfreq(window_size, 1 / 200)
    new_data = 2 * fft_data / len(fft_data)
    new_data = np.abs(new_data) ** 2
    vlf_a = 0.0001
    lf_a = 0.04 # Нижняя граница частот в Гц
    lf_b = 0.15 # Верхняя граница частот в Гц
    hf_b = 0.4
    vlf_idxs = np.where((f >= vlf_a) & (f < lf_a))[0]
    lf_idxs = np.where((f >= lf_a) & (f < lf_b))[0]
    hf_idxs = np.where((f >= lf_b) & (f <= hf_b))[0]
    mean_VLF_power = np.mean(new_data[vlf_idxs])
    mean_LF_power = np.mean(new_data[lf_idxs])
    mean_HF_power = np.mean(new_data[hf_idxs])
    LF_to_HF = mean_LF_power / mean_HF_power
    hf_arr = new_data[hf_idxs]
    hf_freq_arr = f[hf_idxs]
    max_hf_indx = np.argmax(hf_arr)
    max_hf_value = hf_arr[max_hf_indx]
    max_hf = hf_freq_arr[max_hf_indx]
    max_hf_a = max_hf - 0.05
    max_hf_b = max_hf + 0.05
    apr_max_hf_idxs = np.where((hf_freq_arr >= max_hf_a) & (hf_freq_arr <= max_hf_b))[0]
    mean_apr_hf_power = np.mean(hf_arr[apr_max_hf_idxs])
    stats.append([mean_VLF_power, mean_LF_power, mean_HF_power, LF_to_HF, max_hf_value, mean_apr_hf_power])

  return np.array(stats)
```

```python
def progressive_detrended_fluctuation_analysis(rr_intervals, time_samples, window_size):

    n = len(rr_intervals)
    profile = np.cumsum(rr_intervals - np.mean(rr_intervals))

    num_windows = n // window_size
    F = []
    trends = []
    P = []
    angles = []
    for i in range(num_windows):
        start = i * window_size
        end = start + window_size
        segment = profile[start:end]

        x = np.arange(window_size)
        coeffs = np.polyfit(x, segment, 1)
        trend = np.polyval(coeffs, x)
        trends.extend(trend)


    N = len(profile)
    s = 0
    for i in range(num_windows):
        start = i * window_size
        end = start + window_size
        segment_profile = profile[start:end]
        segment_trend = trends[start:end]
        #s = 0
        l_window = len(segment_profile)
        for p in range(l_window):
          s += (segment_profile[p] - segment_trend[p]) ** 2
          P.append(np.sqrt(s))

    P_len = len(P)
    for i in range(1, P_len - 2):
      angle = (P[i + 1] - P[i - 1]) / (time_samples[i + 1] - time_samples[i - 1]) #Так как мы работаем с неинтерполированными RRми, то наш шаг будет неравномерным, не будет ли точность скакать?
      angles.append(angle)

    return angles, P
```

```python
def calculate_mean_pdfa(data, pdfa, start_time=525, window_size=270, step=30):
  stats = []
  window_half = window_size // 2

  for window_start in range(start_time, len(data), step):
    window_end = window_start + window_size
    window_indices = np.where((data[:, 0] >= window_start - window_half) &
                             (data[:, 0] <= window_start + window_half))[0]
    #print(window_indices)
    if len(window_indices) > 0:
      angles = pdfa[window_indices[0]: window_indices[-1]]
      mean_angles = np.mean(angles)
      stats.append(mean_angles)

  return stats
```

```python
def create_dataset(data, data_inter, annot, window_size, step_size, fs_rr, fs_annot):
    detrended_data = detrend_rr(data)
    print(1)

    stats_orig = calculate_stats(data)
    hrv_orig = calculate_hrv_features(data)
    percentiles_orig = calculate_percentile(data)
    freq_orig = calculate_freq_features(data_inter)
    print(2)

    dfa, P = progressive_detrended_fluctuation_analysis(data[:, 1], data[:, 0], 64)
    mean_pdfa = calculate_mean_pdfa(data, dfa)
    print(3)

    stats_detr = calculate_stats(detrended_data)
    hrv_detr = calculate_hrv_features(detrended_data)
    percentiles_detr = calculate_percentile(detrended_data)
    print(4)

    num_windows = min(
        len(stats_orig), len(hrv_orig), len(percentiles_orig),
        len(freq_orig), len(mean_pdfa),
        len(stats_detr), len(hrv_detr), len(percentiles_detr)
    )
    print(num_windows)
    print(5)

    win_samples = int(window_size * fs_rr)
    step_samples = int(step_size * fs_rr)
    labels = []
    num_annot = len(annot)

    #for i in range(17, num_annot):
      #labels.append(annot[i])
    #labels = labels[:num_windows]
    print(6)
    def truncate(x): return x[:num_windows]

    stats_orig = truncate(stats_orig)
    print(7)
    hrv_orig = truncate(hrv_orig)
    print(8)
    percentiles_orig = truncate(percentiles_orig)
    print(9)
    freq_orig = truncate(freq_orig)
    print(10)
    freq_orig = freq_orig[:num_windows]
    mean_pdfa = truncate(mean_pdfa)
    print(11)
    stats_detr = truncate(stats_detr)
    print(12)
    hrv_detr = truncate(hrv_detr)
    print(13)
    percentiles_detr = truncate(percentiles_detr)


    mean_pdfa = np.array(mean_pdfa)
    if mean_pdfa.ndim == 1:
        mean_pdfa = mean_pdfa.reshape(-1, 1)
    print(14)
    raw_windows = []
    fs = 200
    window_size = 270
    len_rr = len(data_inter)
    window_size *= fs
    window_half = window_size // 2
    begin = 0
    end = 0
    start_time = 105000
    step = 30 * fs
    count = 17
    array_drop = []
    windows_center = []
    for window_start in range(start_time, len_rr, step):
      windows_center.append(window_start // 200)
      window_end = window_start + window_size
      begin = start_time - window_half
      end = start_time + window_half
      rr_intervals = data_inter[begin:end]
      st = start_time / 200
      wh = window_half // 200
      window_indices = np.where((data[:, 0] >= st - wh) &
                             (data[:, 0] <= st + wh))[0]
    #print(window_indices)
      if len(window_indices) > 0:
        raw_windows.append(rr_intervals)
      else:
        array_drop.append(count)
      count += 1
    print(15)
    for i in range(17, num_annot):
      #if annot[i] not in array_drop:
      labels.append(annot[i])
    print(16)
    num_l = len(labels)
    labels = labels[:num_windows]
    windows_center = windows_center[:num_windows]
    print(17)
    raw_windows = raw_windows[:num_windows]
    def truncate_l(x): return x[:num_l]
    if num_l < num_windows:
      stats_orig = truncate_l(stats_orig)
      hrv_orig = truncate_l(hrv_orig)
      percentiles_orig = truncate_l(percentiles_orig)
      freq_orig = truncate_l(freq_orig)
      #freq_orig = freq_orig[:num_windows]
      mean_pdfa = truncate_l(mean_pdfa)
      stats_detr = truncate_l(stats_detr)
      hrv_detr = truncate_l(hrv_detr)
      percentiles_detr = truncate_l(percentiles_detr)
      windows_center = truncate_l(windows_center)
    print(19)
    if num_l < len(raw_windows):
      raw_windows = raw_windows[:num_l]
    print(19)
    windows_center = np.array(windows_center)
    if windows_center.ndim == 1:
        windows_center = windows_center.reshape(-1, 1)
    features = np.hstack([
        windows_center, stats_orig, hrv_orig, percentiles_orig, freq_orig,
        stats_detr, hrv_detr, percentiles_detr,
        mean_pdfa
    ])

    print(20)
    feature_names = (
        ["centers", "mean", "median"] +
        ["sdnn", "rr_range", "pnn50", "rmssd", "sdsd", "mad"] +
        ["perc_5", "perc_10", "perc_25", "perc_50", "perc_75", "perc_90", "perc_95"] +
        ["vlf", "lf", "hf", "lf_hf_ratio", "hf_max", "hf_avg_near_max"] +

        ["mean_detr", "median_detr"] +
        ["sdnn_detr", "rr_range_detr", "pnn50_detr", "rmssd_detr", "sdsd_detr", "mad_detr"] +
        ["perc_5_detr", "perc_10_detr", "perc_25_detr", "perc_50_detr", "perc_75_detr", "perc_90_detr", "perc_95_detr"] +

        ["mean_pdfa"]
    )
    print(21)

    df = pd.DataFrame(features, columns=feature_names)
    df["label"] = labels
    #df["raw_signal"] = raw_windows




    return df
```
